```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#importing dataset
df = pd.read_csv('breast-cancer-wisconsin.data',delimiter=",")
```

```python
df.columns = ['Sample code number', 'Clump Thickness', 'Uniformity of Cell Size',' Uniformity of Cell Shape','Marginal Adhesion','Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin','Normal Nucleoli','Mitoses','Class']
df
```

# EDA

```python
#checking nulls
df.isnull().sum()
```

```python
#checking duplicates
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

We converted all '?' values to NULL, then we created a different datafram 'df_clean' which contained all non-null values and converted it to integers.Then, we will us 'df_clean' to find out the mean and subsitute the value in original dataframe 'df' where value=nan

```python
#look for ? values in Bare Nuclei column in df and replace it with null in the dataset without making a copy
df['Bare Nuclei'].replace('?',np.nan,inplace=True) 
```

```python
#creating a new DataFrame (df_clean) containing only the rows from the "Bare Nuclei" column that have valid, non-missing values.
df_clean = df[['Bare Nuclei']].dropna()
```

```python
#converting the data type of the "Bare Nuclei" column in df_clean to integers, preparing it for numerical analysis.
df_clean = df_clean.astype('int')
```

```python
df['Bare Nuclei'] = df['Bare Nuclei'].fillna(df_clean.mean()) 
```

```python
df['Bare Nuclei'].isna().sum()
```

```python
df=df.dropna()
```

```python
df.info()
```

```python
df['Bare Nuclei'] = df['Bare Nuclei'].astype(int)
```

```python
df.info()
```

# Logistic Model

```python
#splitting independent and dependent features
X = df.drop('Class',axis=1)
Y = df['Class'].copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {'penalty': ['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,50,60],'max_iter':[100,200,300,500]}
```

```python
classifier_regressor = GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=10) #cv = cross validation
```

```python
classifier_regressor.fit(x_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
#prediction
y_pred = classifier_regressor.predict(x_test)
```

```python
#accuracy
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_test,y_pred)
```

```python
score
```

```python
report = classification_report(y_test,y_pred)
```
