```python
#importing Libraries
import pandas as pd
import numpy as np
```

```python
#loading datatset
df = pd.read_csv('customer_purchase_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#splitting dataset
X = df.drop('PurchaseStatus',axis=1)
Y = df['PurchaseStatus'].copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,50,60],'max_iter':[100,200,300,500]}
```

```python
classifier_regressor = GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=10) #cv = cross validation
```

```python
classifier_regressor.fit(x_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
#prediction
y_pred = classifier_regressor.predict(x_test)
```

```python
#accuracy
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_test,y_pred)
```

```python
score
```

```python
report = classification_report(y_test,y_pred)
```
