```python
#handling missing value in Age
def impute_age(df):
   for pclass in df['Pclass'].unique():
       median_age = df[df['Pclass'] == pclass]['Age'].median()
       df.loc[(df['Age'].isnull()) & (df['Pclass'] == pclass), 'Age'] = median_age
   return df
```

```python
df = impute_age(df)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Sex'] = df['Sex'].map({'male':0,'female':1})
```

```python
df
```

```python
#splitting dataset into independent and dependent features
feature_cols = ['Pclass', 'Sex', 'Age', 'Fare', 'SibSp', 'Parch']
X = df[feature_cols]
Y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300,400]}
```

```python
classifier_regressor = GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=10) #cv = cross validation
```

```python
classifier_regressor.fit(x_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
y_pred = classifier_regressor.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
score = accuracy_score(y_test,y_pred)
```

```python
score
```

```python
report = classification_report(y_test,y_pred)
```
